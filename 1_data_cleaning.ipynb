```python
df_discogs = pd.read_csv("data/discogs.csv")
```

```python
df_discogs = df_discogs.drop(columns=['notes', 'video_url', 'company_name', 'status', 'id',])
df_discogs
```

```python
df_discogs = df_discogs.drop(columns=['master_id'])
```

```python
df_discogs_clean = df_discogs.dropna()
df_discogs_clean
```

```python
df_discogs_clean.dtypes
```

```python
df_discogs_clean['release_year'] = df_discogs_clean['release_date'].astype('int64')
```

```python
df_discogs_clean = df_discogs_clean.drop(columns=['release_date'])
```

```python
df_discogs_clean = df_discogs_clean.sort_values(by='release_year', ascending=True)
```

```python
num_rows_with_zero = df_discogs_clean['release_year'].value_counts().get(1940, 1940)
print(f"Number of rows with 0 in 'release_year': {num_rows_with_zero}")
```

Now i want to see which is the first electronic release, and then i am going to delete all the rows that have release year before that.

```python
df_clean_electronic = df_discogs_clean[df_discogs_clean['genre'] == 'Electronic']
```

```python
df_clean_electronic.head(15)
```

```python
num_rows_with_zero = df_clean_electronic['release_year'].value_counts().get(0, 0)
print(f"Number of rows with 0 in 'release_year': {num_rows_with_zero}")
```

now i am going to get rid of all the rows with the release_year less than 1950, seeing that there is only a couple of releases before that year, and the first releases on vinyl are from the year 1950, and also, around that year were the first musique concrete releases, which is generally considered the beginning of modern electronical music

```python
mask = df_discogs_clean['release_year'] >= 1950

df_discogs_clean = df_discogs_clean[mask]
```

```python
df_discogs_clean
```

```python
df_discogs_clean.to_csv('data/discogs_clean.csv', index=False)
```
